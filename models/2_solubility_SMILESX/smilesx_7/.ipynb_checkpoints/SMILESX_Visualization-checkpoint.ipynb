#### Definition of data

```python
validation_data_dir = "./input_dataset/"
```

```python
!ls -l $validation_data_dir
```

```python
data_name = 'solubility_total' # 'delaney'
```

```python
extension = '.csv'
```

```python
# read the dataset
sol_data = pd.read_csv(validation_data_dir + 'solubility_total' + extension)
```

```python
sol_data.info()
```

```python
sol_data.head(3)
```

```python
# rename the name columns to comply with the SMILESX package:
# the column with the SMILES structures has to have the name 'smiles'
sol_data.columns = ['smiles','exp_logS']
```

```python
sol_data.head(3)
```

```python
# check the size of the dataset
sol_data.shape
```

#### Embedding Visualization

```python
embeddingvis.Embedding_Vis(data=sol_data, 
                          data_name=data_name, 
                          data_units='', 
                          k_fold_number = 1,
                          k_fold_index = 0,
                          augmentation = True, 
                          outdir = "./data/", 
                          affinity_propn = True)
```

#### Find tokens in SMILES

```python
token.TokensFinder(data=sol_data, 
                   data_name=data_name, 
                   data_units='', 
                   k_fold_number = 1,
                   k_fold_index = 0,
                   augmentation = True, 
                   token_tofind = '[nH]', 
                   verbose = 0)
```

#### SMILES inference interpretation

```python
template_test = 'CCCN1N=C(C(=O)N2CCCCC2)C2=C1CCC(C2)NCCN1CCCCC1' # CN_S003_C005
#template_test = 'Cc1ccc(O)cc1C'
Chem.MolFromSmiles(template_test)
```

```python
interpret.Interpretation(data=sol_data, 
                         data_name=data_name, 
                         data_units='', 
                         k_fold_number = 1,
                         k_fold_index = 0,
                         augmentation = True, 
                         outdir = "./data/", 
                         smiles_toviz = template_test, 
                         font_size = 15, 
                         font_rotation = 'vertical')
```

```python
interpret.Interpretation(data=sol_data, 
                         data_name=data_name, 
                         data_units='', 
                         k_fold_number = 1,
                         k_fold_index = 0,
                         augmentation = True, 
                         outdir = "./data/", 
                         smiles_toviz = 'c1ccsc1', 
                         font_size = 15, 
                         font_rotation = 'vertical')
```

```python
interpret.Interpretation(data=sol_data, 
                         data_name=data_name, 
                         data_units='', 
                         k_fold_number = 1,
                         k_fold_index = 0,
                         augmentation = True, 
                         outdir = "./data/", 
                         smiles_toviz = 'CC(C)CCOC1=CC2=C(C=C1)C1=CC=NC(C)=C1N2', 
                         font_size = 15, 
                         font_rotation = 'vertical')
```
